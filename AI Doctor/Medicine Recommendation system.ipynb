# Load dataset

```python
import pandas as pd
import numpy as np
```

```python
ds=pd.read_csv(r'D:\Road to ML\datasets\Training.csv')
```

# Encode Categorical column using label encoder

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
ds_new=ds
```

```python
ds_new['prognosis']=LE.fit_transform(ds['prognosis'])
```

# Apply train test split

```python
from sklearn.model_selection import train_test_split

x=ds_new.drop('prognosis',axis=1)
y=ds_new['prognosis']

X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3,random_state=42)

X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

# Train Top Classification models

```python
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier


from sklearn.metrics import accuracy_score,confusion_matrix


models={
    'RFC':RandomForestClassifier(n_estimators=100,random_state=42),
    'GBC':GradientBoostingClassifier(n_estimators=100,random_state=42),
    'svc':SVC(kernel='linear'),
    'NB':MultinomialNB(),
    'KNC':KNeighborsClassifier(n_neighbors=5)
}

for name,model in models.items():
    model.fit(X_train,Y_train)
    y_pred=model.predict(X_test)


    print(f"{name} Consuion_matrix: ")
    print(confusion_matrix(Y_test,y_pred))
    print(f"{name} Accuracy is: ")
    print(accuracy_score(Y_test,y_pred))
```

# Select best model and make some prediction for testing

```python
svc=SVC()
svc.fit(X_train,Y_train)
y_pred=svc.predict(X_test)
print("Accuracy is: ")
print(accuracy_score(Y_test,y_pred))



```

```python
print("Test 1")
print("Predicted label: ",svc.predict(X_test.iloc[60].values.reshape(1,-1)))
print("Actual label: ",Y_test.iloc[60])
print("Test 2")
print("Predicted label: ",svc.predict(X_test.iloc[17].values.reshape(1,-1)))
print("Actual label: ",Y_test.iloc[17]) 
print("Test 3")
print("Predicted label: ",svc.predict(X_test.iloc[55].values.reshape(1,-1)))
print("Actual label: ",Y_test.iloc[55])
```
